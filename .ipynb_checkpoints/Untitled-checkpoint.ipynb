```python
%matplotlib notebook
from numpy import *
from pylab import *
import matplotlib.pyplot as plt 
import matplotlib.image as mgimg
from matplotlib import animation

## Read in graphs

p = 0
myimages = []
fig = plt.figure()
nbImages = 39
for k in range(0, nbImages):
    fname = "frame" + str(k)+".png"
      # read in pictures
    img = mgimg.imread(fname)
    
    if k==0:
        im = plt.imshow(img,animated=True)
        #print(img[:,:,0])
        
    #print(np.shape(img))
    myimages.append(img)
    

def updatefig(i):
    
    im.set_array(myimages[i])
    return im,

anim = animation.FuncAnimation(fig, updatefig ,frames=nbImages)

plt.show()
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure()


def f(x, y):
    return np.sin(x) + np.cos(y)

x = np.linspace(0, 2 * np.pi, 120)
y = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)
print(np.shape(f(x, y)))
im = plt.imshow(f(x, y), animated=True)


def updatefig(*args):
    global x, y
    x += np.pi / 15.
    y += np.pi / 20.
    print(f(x, y))
    im.set_array(f(x, y))
    return im,

ani = animation.FuncAnimation(fig, updatefig,frames=100, interval=50, blit=True)
plt.show()
```
